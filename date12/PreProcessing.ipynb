```python
# COVID-19데이터 활용 EDA
# Province, Gender, Time 세 개의 csv 파일을 가지고 시작.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# pandas를 이용해서 정형화된 데이터 가져오기.
Time = pd.read_csv('C:/Users/User/Desktop/Time.csv')
# 해당 데이터의컬럼 형식, 갯수, null유무 파악
Time
```

```python
TimeGender = pd.read_csv('C:/Users/User/Desktop/TimeGender.csv')
TimeGender
```

```python
TimeProvince = pd.read_csv('C:/Users/User/Desktop/TimeProvince.csv')
# .groupby('date').mean()
TimeProvince
```

```python
TimeAge = pd.read_csv('C:/Users/User/Desktop/TimeAge.csv')
# .groupby('date')
TimeAge
```

```python
Time_gender = Time.loc[Time.index.repeat(2)].reset_index(drop=True)
Gender_ratio = TimeGender['confirmed'] / Time_gender['confirmed']
Gender_ratio * 100
```

```python
Time_age = Time.loc[Time.index.repeat(9)].reset_index(drop=True)
Age_ratio = TimeAge['confirmed'] / Time_age['confirmed']
Age_ratio = Age_ratio * 100
Age_ratio
```

```python
Time_province = Time.loc[Time.index.repeat(17)].reset_index(drop=True)
Province_ratio = TimeProvince['confirmed'] / Time_province['confirmed']
Province_ratio
```

```python
total_ratio = Gender_ratio * Age_ratio * Province_ratio
total_ratio
```

```python
Time.corr() # 항목을 지정하지 않았으므로, 각 항목들의 상관도를 구해줌(-1~1)
```

```python
sns.heatmap(Time.corr(), linewidths=0.01, square=True,
           annot=True, cmap=plt.cm.viridis, linecolor="white")
plt.title('Correlation between features')
plt.show()
```
